```python
import pandas as pd
import os

# --- Define Possible Paths ---
# Check the 'runs/detect/' folder to determine the correct path.
# Change this variable to 'train' if that is your latest run.
RUN_FOLDER = "runs/detect/train"
results_path = os.path.join(RUN_FOLDER, "results.csv")

# 1. Load the CSV file
try:
    results = pd.read_csv(results_path, skipinitialspace=True)
except FileNotFoundError:
    print(f"--- ERROR: File Not Found ---")
    print(f"The path '{results_path}' does not exist.")
    print("Please check your 'runs/detect/' folder and update the RUN_FOLDER variable to 'train' or 'train2'.")
    # Exit or handle the error gracefully
    raise # Re-raise the error to stop execution if the path is still wrong

# 2. Calculate the official YOLO Fitness Score (as the "Highest Percentage")
map50 = results.iloc[:, 6]    # metrics/mAP50(B)
map50_95 = results.iloc[:, 7] # metrics/mAP50-95(B)
fitness = (0.1 * map50) + (0.9 * map50_95)

# 3. Find the best epoch
best_epoch_index = fitness.idxmax()
highest_percentage_epoch = best_epoch_index + 1
best_fitness_score = fitness.max()
best_mAP50_95 = map50_95.iloc[best_epoch_index]


print(f"--- Analysis Complete (From {RUN_FOLDER}) ---")
print(f"The epoch with the highest performance (Fitness) is: **Epoch {highest_percentage_epoch}**")
print(f"The corresponding fitness score is: **{best_fitness_score:.4f}**")
print(f"The mAP@0.5:0.95 (B) for this epoch is: {best_mAP50_95:.4f}")
```
